```python
import numpy as np
samples = np.random.normal(loc=0.0, scale=1.0, size=1000)
plt.figure(figsize=(6, 4))
plt.hist(samples, bins=30, density=True, alpha=0.6, color='b')
plt.title('Histogram of Standard Normal Samples')
plt.xlabel('Value')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```
